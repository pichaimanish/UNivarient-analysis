```python
import numpy as np
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#,Data,directory
def graph(Directory,Data,directory,c,desp):        # graph declaration  and arguments
    if(desp=='y'):
        Data.describe()                   ## describe of data set
        Data.info()
        #Data.shape()

        cars=Data.copy()      # here file will make a copy  
        
        col=cars.columns
        cl=cars.shape[1]         # c1 store shape of car
        for i in range(0,cl):     # using loop to iterate over column
            if cars[col[i]].dtypes=='O' and len(cars[col[i]].unique())<cars.shape[0]/2:

                cars[col[i]].value_counts().plot(kind='barh',color=c)
                plt.title(col[i])
                if not os.path.exists(directory): ##if directory not present
                    os.makedirs(directory)
                    path=os.getcwd()+directory
                    os.chdir(path)  ##this will take path of our field

                    plt.savefig(col[i]+'.png')   ## it will save your picture
                    os.chdir(Directory)
                    plt.show()


            elif cars[col[i]].dtypes=='int64' or cars[col[i]].dtypes=='float64': 

                plt.hist(x=cars[col[i]],data=cars)
                plt.title(col[i])
                plt.show()
                if not os.path.exists(directory):
                    os.makedirs(directory)
                    path=os.getcwd()+directory
                    os.chdir(path)
                    plt.savefig(col[i]+'.png')
                    os.chdir(Directory)
                    plt.show()

                plt.boxplot(x=cars[col[i]])
                plt.title(col[i])
                plt.savefig(col[i]+'.png')
                plt.show()
            


        file1 = open("MyFile.txt","a") 
        file1 = open("myfile.txt","w")
        file1.write(str(Data.corr()))
        file1.write('Correlation details')
        file1.write(str(Data.describe()))
        #file1.writelines(Data.info()) 
        file1.close()
        file1 = open("myfile.txt","r+")
        os.getcwd()
        
        
    else:
        
            
        cars=Data.copy()
        cars.head()
        col=cars.columns
        cl=cars.shape[1]
        for i in range(0,cl):
            if cars[col[i]].dtypes=='O' and len(cars[col[i]].unique())<cars.shape[0]/2: 
            #checking the column is categorical or not and give bar graph if the unique value is less than the total column data

                cars[col[i]].value_counts().plot(kind='barh',color=c)
                plt.title(col[i]) #Plot the graph
                if not os.path.exists(directory):
                    os.makedirs(directory)
                    path=os.getcwd()+directory #change the path
                    os.chdir(path)

                    plt.savefig(col[i]+'.png')
                    os.chdir(Directory)
                    plt.show()


            elif cars[col[i]].dtypes=='int64' or cars[col[i]].dtypes=='float64': #checking the column is numerical or not

                plt.hist(x=cars[col[i]],data=cars)
                plt.title(col[i]) #Giving the title
                plt.show()
                if not os.path.exists(directory): #If path not exist
                    os.makedirs(directory) #Create directory
                    path=os.getcwd()+directory #change directory path
                    os.chdir(path)
                    plt.savefig(col[i]+'.png') #save image in png
                    os.chdir(Directory)
                    plt.show() #show all pic

                plt.boxplot(x=cars[col[i]])
                plt.title(col[i])
                plt.savefig(col[i]+'.png') #save in png format
                plt.show() #it will show the graph
```
